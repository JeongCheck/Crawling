# Naver 검색 API 를 사용해 크롤링한 데이터 검증하기 위한 파일
- 글이 중복되는 문제가 발생해 크롤링한 데이터를 가져와 확인한다.

```python
import os
import pandas as pd
import numpy as np

data_dir = os.getenv('HOME') + '/Projects/crawling_data'
data_dir
```

```python
file_list = os.listdir(data_dir)
print(len(file_list))
file_list
```

```python
def check_data(file_name):
    read_data = pd.read_csv(data_dir+f'/{file_name}', names=['date', 'desc', 'text', 'name'])
    print('검색 키워드명 :', file_name)
    print('파일 개수 :', len(read_data))
    print('블로그 명 개수 :', len(set(read_data.name)))
    
    if len(read_data) != len(set(read_data.name)):
        print(read_data.head())
```

```python
check_data('오월드여행대전.csv')
```

```python
check_data('오월드 여행 대전.csv')
```

```python
for file in file_list:
    check_data(file)
```
